```python
# Define a DNA sequence:
seq = "ATGCGAATTCGCTAG"
```

### Strings as a sequence of characters

In Python we locate the position of a character in a string using a system called indexing. The most critical rule to rememberâ€”is that **Python is a zero-index-based language**. This means the very first character in a sequence is at index 0, the second is at index 1, and so on. If you want to extract a specific character, you use square brackets like `seq[0]`. This logic also applies to slicing, where you extract a range of characters. For example, `seq[0:3]` will give you the first three bases (the start codon), effectively grabbing the characters at positions 0, 1, and 2, but **stopping before** it reaches index 3.

```python
print("The first base in the sequence is: ",seq[0])
print("The last base in the sequence is: ",seq[-1])
```

### âœ‚ï¸ Sequence Indexing & Slicing Reference

| Python Syntax | Biological Meaning | Example (for `seq = "ATGCGT"`) |
| :--- | :--- | :--- |
| **`seq[0]`** | The **first** base | `"A"` |
| **`seq[1]`** | The **second** base | `"T"` |
| **`seq[-1]`** | The **very last** base | `"T"` |
| **`seq[-2]`** | The second to last base | `"G"` |
| **`seq[0:3]`** | The **first codon** (indices 0, 1, 2) | `"ATG"` |
| **`seq[3:6]`** | The **second codon** (indices 3, 4, 5) | `"CGT"` |
| **`seq[:]`** | The **entire** sequence | `"ATGCGT"` |

Since **DNA**, **RNA**, and **Proteins** are represented as strings, these methods allow you to perform *"digital biochemistry"* like finding **motifs**, measuring length, or cleaning up messy data from a sequencer.

Here are the most useful methods:



----

**`len()`** â€” **Sequence Length**

```python
len(seq)
```

---

**`.count()`** â€” **Frequency Analysis**

```python
seq.count('A')
```

---

**`.lower()`, `.upper()` & `.capitalize()`**  - **Convert characters to lower or upper case**

```python
fruit_fly = "dRosophila (commonly fruit flies) are a genus of small flies."

print( fruit_fly.lower() )
print( fruit_fly.upper() )
print( fruit_fly.capitalize() )
```

---

**`.startswith()`** and **`.endswith()`** â€” **Quality Control**

Excellent for checking for specific start codons (ATG) or poly-A tails.

```python
mrna = "AUGCCCCAAAAAAAAA"

# Check if the mRNA sequence ends with poly-A tail.
print(mrna.endswith("AAAAA"))

if mrna.endswith("AAAAA"):
    print("Poly-A tail detected.")
```


---

**`.find()`** and **`.index()`** â€” **Motif Localization**

These methods help you find the index ( or position) of a specific sub-sequence.

`.find()` returns -1 if the motif isn't found.

`.index()` throws an error if not found.

```python
# The .find() method is "gentle." If it cannot find the substring you are looking for, it returns -1.
seq1 = "ATGCCGATGCCG" # No 'GAATTC' site here
seq2 = "ATGCCGAATTCCG" # 'GAATTC' site is present here

# Find position of GAATTC site in each sequence
pos1 = seq1.find('GAATTC')
pos2 = seq2.find('GAATTC')

if pos1 == -1:
    print("Enzyme", pos1, "does not cut this sequence.")
else:
    print(f"Cutting {seq1} at position {pos1}")


if pos2 == -1:
    print("Enzyme", pos2, "does not cut this sequence.")
else:
    print(f"Cutting {seq2} at position {pos2}")

```

```python
# The following will produce an error:
seq1.index('GAATTC')
```

---

**`.replace()`** â€” **Transcription and Mutagenesis**

Used to convert DNA to RNA (T to U) or to simulate a point mutation in a sequence:

```python
dna = "ATGCATGCCGATGCCG"
rna = dna.replace("T", "U")
print(rna)
```

---

**`.strip()`** â€” **Cleaning Raw Data**

When you read data from a file (like a **FASTA**), it often includes invisible *"newline"* characters (\n) or extra spaces. `.strip()` removes these from the start and end.

```python
raw_data = "  ATATGCATGCCGATGCCGGCGTA\n"
print(f"Raw String: {raw_data}")

clean_data = raw_data.strip()
print(f"'{clean_data}'")
```
